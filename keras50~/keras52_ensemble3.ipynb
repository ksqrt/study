```python
import numpy as np
#  아모레 시가 ,고가
x1_datasets = np.array([range(100),range(301,401)]).transpose()
x1_datasets.shape # (100,2)
```

```python
# 삼성전자의 시가, 고가 ,종가
x2_datasets = np.array([range(101,201),range(411,511),range(150,250)]).T
x2_datasets.shape # (100,2)
```

```python

x3_datasets = np.array([range(100,200),range(1301,1401)]).T
x3_datasets.shape # (100,2)
```

```python
# 삼성전자의 하루 뒤 종가
y1 = np.array(range(2001,2101))
y2 = np.array(range(201,301))
y1.shape
```

```python
from sklearn.model_selection import train_test_split
x1_train,x1_test,x2_train,x2_test,x3_train,x3_test,y1_train,y1_test,y2_train,y2_test= train_test_split(
    x1_datasets,x2_datasets,x3_datasets,y1,y2,
    train_size = 0.7
)
```

```python
print(x1_train.shape,x2_train.shape,x3_train.shape,y1_train.shape,y2_train.shape)
```

```python
from tensorflow.keras.models import Sequential,Model
from tensorflow.keras.layers import Dense,Input

# 모델 1
input1 = Input(shape=(2,))
dense1 = Dense(64,activation="relu",name="d1")(input1)
dense2 = Dense(32,activation="relu",name="d2")(dense1)
dense3 = Dense(16,activation="relu",name="d3")(dense2)
output1 = Dense(8,activation="relu",name="d4")(dense3)

# 모델 2
input2 = Input(shape=(3,))
dense11 = Dense(64,activation="relu",name="d11")(input2)
dense12 = Dense(32,activation="relu",name="d12")(dense11)
dense13 = Dense(16,activation="relu",name="d13")(dense12)
output11 = Dense(8,activation="relu",name="d14")(dense13)

# 모델 3
input21 = Input(shape=(2,))
dense21 = Dense(64,activation="relu",name="d21")(input21)
dense22 = Dense(32,activation="relu",name="d22")(dense21)
dense23 = Dense(16,activation="relu",name="d23")(dense22)
output21 = Dense(8,activation="relu",name="d24")(dense23)

```

```python
# 모델 병합
from tensorflow.keras.layers import concatenate

merge1 = concatenate([output1,output11,output21],name ="mg1")
merge2 = Dense(4,activation="relu",name="mg2")(merge1)
merge3 = Dense(1,name ="mg3")(merge2)
```

```python
# 모델 분기1
dense31 = Dense(64,activation="relu",name="d121")(merge3)
dense32 = Dense(32,activation="relu",name="d122")(dense31)
dense33 = Dense(16,activation="relu",name="d123")(dense32)
output31 = Dense(1,activation="relu",name="d124")(dense33)

```

```python
# 분기2
dense41 = Dense(64,activation="relu",name="d221")(merge3)
dense42 = Dense(32,activation="relu",name="d222")(dense41)
dense43 = Dense(16,activation="relu",name="d223")(dense42)
output41 = Dense(1,activation="relu",name="d224")(dense43)
```

```python
model = Model(inputs=[input1,input2,input21],outputs=[output31,output41])
model.summary()
```

```python
# 3. 컴파일 훈련
model.compile(loss="mse",optimizer="adam")
model.fit([x1_train,x2_train,x3_train],[y1_train,y2_train],epochs =200 ,batch_size=8)
```
