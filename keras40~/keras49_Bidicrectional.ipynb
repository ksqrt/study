```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,SimpleRNN,LSTM,GRU,Dropout,Bidirectional


a = np.array(range(1,101))
x_predict = np.array(range(96,106))

```

```python
def split_x(dataset,timesteps) : 
    tmp = []
    for i in range(len(dataset)-timesteps + 1):
        subset = dataset[i : (i+ timesteps)]
        tmp.append(subset)
    return np.array(tmp)

bbb = split_x(a,5)

print(bbb.shape)

x = bbb[:,:-1]
y = bbb[:,-1]

print(x,"\n",y)
```

```python
print(x.shape,y.shape)
```

```python
# 모델을 리쉐이프해서 7개 데이터 를 (3,1) 로 만들자
x = x.reshape(len(x),2,2)
print(x.shape)
print(x)
```

```python
print(x.shape,y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(
    x,y, shuffle = True,
    train_size=0.7,

    
)
print(x.shape,y.shape)
```

```python
# RNN 모델구성
model = Sequential([
    Bidirectional(LSTM(64),input_shape=(2,2)),
    # Bidirectional 을 사용하면 일반 LSTM 에 비해 연상량이 2배 증가합니다.
    Dropout(0.2),
    Dense(16,activation="relu"),
    Dense(1)
])

model.summary()
```

```python
# 컴파일 훈련
model.compile(loss="mae",optimizer="adam")
model.fit(x_train,y_train,epochs=100,batch_size=1)
```

```python
loss = model.evaluate(x,y)

print(loss)
```

```python
x_predict = np.array(range(96,106))
x_predict = split_x(x_predict,4)
# x_predict.shape
x_predict = x_predict.reshape(7,2,2)
x_predict.shape
# (7,4) - > (7,4,1)
result = model.predict(x_predict)
print("결과 \n",result)
```

```python
test = np.array([[[96, 97],[98, 99]]])

# print(test.shape)

result = model.predict(test)
print("결과",result)
```
